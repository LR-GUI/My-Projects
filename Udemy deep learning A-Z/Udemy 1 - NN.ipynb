We will not use the noninformative columns: RowNumber, CustomerId, and Surname.

```python
X = df.iloc[:,3:13].values # Removed columns are the first three
y = df.iloc[:,13].values # The column 13 is the target
X.shape,y.shape
```

Now we numerically encode the categorical variables: Geography and Gender.

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# LabelEncoder transforms string categories into integer categories

le1 = LabelEncoder()
le2 = LabelEncoder()
X[:,1] = le1.fit_transform(X[:,1])
X[:,2] = le2.fit_transform(X[:,2])
print(pd.DataFrame(X).head(10))
```

```python
# OneHotEncoder removes sequentiality by separating categories into different columns

ohe = OneHotEncoder(categorical_features=[1])
X = ohe.fit_transform(X).toarray()[:,1:]
print(pd.DataFrame(X).head(10))
```

We separate the samples for training and testing:

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

We use StandardScaler to transform the data to data from a gaussian distribution with zero mean and variance = 1. This is important because neural networks are scale sensitive.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(pd.DataFrame(X_train).head(10))
```

Now, we build the model with keras: an input layer with 11 units, one for each feature; two hidden layers with 6 units (mean between input and output layers); and an output layer with 1 unit (giving the probability of exit).

Activation functions are adjusted accordingly: relu for nonlinarization of the model and softmax for an output satisfying probability criteria.

We backpropagate with stochastic gradient descent using adam.

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

classifier = Sequential()
classifier.add(Dense(units=6,input_shape=(11,),activation='relu',kernel_initializer='uniform'))

classifier.add(Dense(units=6,activation='relu',kernel_initializer='uniform'))

classifier.add(Dense(units=1,activation='sigmoid',kernel_initializer='uniform'))

classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
classifier.fit(X_train,y_train,batch_size=10,epochs=10)
```

A first prediction model gives:

```python
y_pred = classifier.predict(X_test)
print(np.around(y_pred,2))
```

```python
from sklearn.metrics import confusion_matrix

y_pred = (y_pred > 0.5) # We use a decision criterion of equal probability for classes
cm = confusion_matrix(y_test,y_pred)
cm
```

With accuracy of 84%.

```python
accuracy = np.trace(cm) / cm.sum()
accuracy
```

Now we perform cross validation to obtain a more meaningful estimate of the model's accuracy:

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score

def nnetwork():
    classifier = Sequential()
    classifier.add(Dense(units=6,input_shape=(11,),activation='relu',kernel_initializer='uniform'))
    classifier.add(Dense(units=6,activation='relu',kernel_initializer='uniform'))
    classifier.add(Dense(units=1,activation='sigmoid',kernel_initializer='uniform'))
    classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
    return classifier

classifier = KerasClassifier(build_fn=nnetwork, batch_size=10, epochs=10)
```

```python
cross = cross_val_score(estimator = classifier, X=X_train, y=y_train, cv=10, n_jobs=1)
```

By this analysis, our model has a more solid 83% accuracy, with low deviation:

```python
cross, cross.mean(), cross.std()
```

Here we implement dropout as regularization (to promote independence between the units) and perform a GridSearch for certain parameter values (the example here is underoptimized due to time reasons):

```python
from keras.layers import Dropout

def reg_nnetwork(optimizer):
    classifier = Sequential()
    classifier.add(Dense(units=6,input_shape=(11,),activation='relu',kernel_initializer='uniform'))
    classifier.add(Dropout(rate=0.1))
    classifier.add(Dense(units=6,activation='relu',kernel_initializer='uniform'))
    classifier.add(Dropout(rate=0.1))
    classifier.add(Dense(units=1,activation='sigmoid',kernel_initializer='uniform'))
    classifier.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
    return classifier
classifier = KerasClassifier(build_fn=reg_nnetwork)

from sklearn.grid_search import GridSearchCV

params = {'batch_size':[10,32], 'epochs':[20], 'optimizer':['adam','rmsprop']}
gs = GridSearchCV(estimator=classifier, param_grid=params, cv=10, scoring='accuracy')
gs.fit(X_train,y_train)
```

Still, we get a consistent 83,5% accuracy with new parameters:

```python
best_params = gs.best_params_
best_score = gs.best_score_
```
