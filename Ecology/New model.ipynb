## New stuff

### Anisogamy
We model anisogamy through a cost in fitness per reproduction event that may be made unequal between sexes. This introduces a paramenter modulating the loss in fitness associated with each offspring, $g$, called reproduction cost. We can talk about an anisogamy degree $G=\frac{g}{\tilde{g}}$, that increases with increasing anisogamy, meaning how many times the female's reproduction cost is larger than the male's. We then have a new fitness function, $E \to W'$:

$W' = W - gbN_m$

Where $bN_m$ is the average number of offspring resulting from all reproduction events scored by the individual. Note that the Fisher condition would still impose $W/\tilde{W}=r$. The modeling of anisogamy is crucial, because this is what separates both populations in the first place, what defines male and female. We could model this indirectly through its consequences (their model does this using as consequences, and separarations for both populations, the parental uncertainty, the strength of sexual selection, and the death rates, along with unexplored variations on the MSR). But these are all parameters, and I propose that the reproduction cost should be the parameter, with the intention of making some of these consequences ($n$, $k$, $r$, $\mu$, that are parameters in their model) dependent variables of parameters.

The idea for now is to change their modeling of $n$ to a more realistic one and to try modeling $k$ as a dependent variable (and potentially changing its definition).

A word about the brood size, $b$. If we choose to model the reproduction cost being proportional to the size of the brood, then it remains irrelevant. But if we choose to instead not increase the cost by the brood size, then we need to drop the $b$ above and it becomes a parameter that can be calculated with the argument of demographic equilibrium.

### Strength of sexual selection

Let's try to make this a dependent variable of other parameters and variables.

The parameter $k$ just modulates the fraction of adults that are qualified to mate, making the rest never able to leave time-in or have offspring (they don't mate). It can be because of intrasexual selection (competition) or intersexual selection (choosiness). Due to the assymetric effects of these two sources between the sexes, we could consider modeling the effects separately, so we can know whether a large $k$ comes from competition or choosiness (so we can be able to say things like this: for this species, represented as a set of parameter values, choosiness evolved in females and competition evolved in both).

- Now, how can we make $k$ depend on our inputs? Obvious candidates to modulate $k$ are the OSR and the ASR, what is the empirical link between them? Should we take more into consideration? How can we differentiate competition from choosiness by the light of the OSR and the ASR?

### Parental uncertainty

The idea is to drop the first two functions of $n$ and correct the third (as explained in the questions to the authors; 1. the entire group goes time-out 2. the entire group cares together 3. everybody caring gains fitness only when is the actual parent). Dropping the first two, we don't have the concept of mating group anymore, and, irrespective of how many males a female copulated with and vice versa, they go to time-out for caring in couples, 1 male and 1 female.

Using just the third function as it is (and it becomes numerically relevant with the introduction of anisogamy), we retain only the meaning of parental uncertainty, decreasing the fitness by a factor of $n$ to say that someone is going to be the actual parent of only $1/n$ of their cared offspring. This raises an issue of missing fitness. Who is gaining the rest of this fitness? Parents that didn't care for offspring. So we must also add a term of increase in fitness due to their actual offspring that was cared by others. Since average care is equal for everyone, what each individual gets is the same fitness they lost with the factor $1/n$, ending up with the same. So parental uncertainty, modeled like this, doesn't alter the average fitness. If the population is homogeneous like this, one can not care less than reproduce, on average, because the total number of reproductions has to match the total number of cares. Unless (!) we can add a chance that someone doesn't want to care for offspring as a consequence of parental uncertainty. This makes possible for individuals to choose not to care in favor of reproducing more, and it sounds much more like a mechanism for parental uncertainty (thinking about it now, although it's irrelevant in computations, their use 3 of $n$ is simply wrong; you can't lose fitness because of parental uncertainty in care without gaining it back, because others made it possible for your offspring in the same way, so you lose to someone and someone loses from you, equally).

- How to best model chance of abandoning care?

### Modeling structure

It is good to understand the model as a tool to represent different types of species behavior, trying to predict their care durations through the specifications given on the parameters set.

If we consider $k$ to be parameter, as in their model, we should have:

- Parameters: these values define the species and are used as mechanisms for the consequences we want to observe
    - reproduction cost ($g$)
    - offspring need of care ($\alpha$)
    - mortality rates ($\mu$)
    - synergy of care ($\gamma$)
    - MSR ($r$)
    - parental uncertainty ($n$)
    - strength of sexual selection ($k$)
    - ******Equation of care
   
   
- Consequences: these are variables defined by the parameter values and the outputs of the model
    - care duration ($T$)
    - OSR ($r_o$)
    - ASR ($r_a$)
    
If we drop the parameter $k$ to model it as consequence, we have it moved from parameter to consequence.


## Update

It seems that anisogamy isn't that important after all, and that's the thing. The Trivers's arguments say that the divergence in parental care comes from anisogamy itself, but it 1) fails to work out a mechanistic explanation (they are just verbal accounts with no rigorous treatment of assumptions) and it 2) fails to account for the various cases where, despite the same anisogamy, parental care evolved in a different manner.

Kokko and Jennions's proposal is to correct these two flaws using a mathematical reasoning and modeling. The first, by using a mathematical framework for trait evolution, based on fitness gradient, that properly works out the assumptions and outputs equilibrium situations for the trait of total care duration. The second, by investigating the different effects and consequenses of anisogamy that themselves explain the different scenarios of parental care we observe. This generates a testable explanation, because the aim is to build a structure that says: given a scenario with such and such evolved consequences of anisogamy (like levels ofcompetitiveness or parental uncertainty), we have this resulting parental care (and it can be verified for species that are under that scenario).

Why anisogamy itself isn't that important? Trivers's account can me summed in two major arguments, what they call the female argument (females have more to lose in abandoning offspring because they already invested more) and the male argument (males have more to lose in staying because they have more prospects in mating). They say that both arguments are flawed, the main joint reason being that they fail to consider all available options males and females have, and their associated impact on fitness.

As a general example, suppose we say that females, because of anisogamy, pay a higher price in fitness to produce gametes. What, in reality, is that higher price in fitness? It has to mean that those females are expected to have less mating success because of that cost (one could argue that females lose in resources become more prone to death, etc, and that affects her overall mating success). But the actual mating success is linked to the mating success of males (this is the Fisher condition, that is pointed as crucial by Kokko and Jennions), and in the model we also should consider this explicitly (not just saying that someone paid a cost and lost fitness, without showing how). We see here that not every cost that is paid ends up affecting fitness (as can be for those of gamete production). Fitness is made by costs and gains explicitly coming from mating success.

So the question becomes: what is generally different between males and females that impacts evolution of parental care? The theory of the model contains three differences that are argued to arise from anisogamy in different ways, depending on environmental contingencies. 1) Sexual selection factors, such as competitiveness and choosiness, expressed by the parameter $k$, 2) parental uncertainty, expressed by the parameter $n$, 3) death rates, expressed by the $\mu$ parameters. Different species have those factors in different tunnings, and the goal is to explain their state and differences of parental care by the differences in those factors. So anisogamy is translated to this consequent complex of differences in other characteristics. For example, depending on how anisogamy participates in a species's evolutionary history, it forces a tendency upon females to become the limiting sex, because of male gametic versatility, thus affecting the $k$ parameter. Now that could properly affect parental care. But then, because of special circumstances, males can be the limiting sex for a specific species, and that may cause this species to have a different parental care, despite coming from the same anisogamy.

Other important factor to consider is the conversation with established empirical population measures, like the OSR and the ASR. In the model, these are not primary factors, but are determined by parameters. Passively as is, they serve as indicators and predictions. But we can further model those anisogamy factors as evolving in mutual feedback with these ratios (like $k$), making them become factors too.

PS. So the thing is tha anisogamy is not important for evolution of parental care. Other things are, and these things can arise from anisogamy combined with other specific factors (even that anisogamy drives the evolution of different stuff, in many occasions, if not all). The real factors of parental care evolution are those that mechanistically drive this evolution. Now, how we can go from anisogamy to the initial differentiation between sexes is shifted to another pursuit in research. This one, of parental care evolution, starts from other factors ($k$, $n$, $\mu$, equation of care, etc), irrespective of anisogamy (that's why it doesn't even show up in the models). Trivers said that the initial anisogamy drives positive feedbacks in divergence of parental care, and that's wrong. There are factors between anisogamy and what happens with parental care. Anisogamy may account for what is most common (because of how those factors arise from it and a typical environment), but does not mechanistically explain what happens with care.

```python
# Deterministic solution function

def Deterministic(step, dt, parameters, initial_state):
    # Calculates the deterministic solution of the model at time steps*dt
    # Inputs the number of time steps covered by the solution, the infinitesimal time dt,
    # the parameters of the model, and the initial state of species
    # Outputs an array of time steps, an array of the solution phi(t=steps*dt), and
    # the deterministic solution function to be used in other computations
    
    # Initialize outputs
    phi0_t = []
    phi1_t = []
    time = [] 
    
    # Model function, with the differential system
    def model(t,y,parameters):
        
        uo = parameters[0]
        ui = parameters[1]
        uo_ = parameters[2]
        ui_ = parameters[3]
        k = parameters[4]
        n = parameters[5]
        M = parameters[6]
        r = parameters[7]
        alpha = parameters[8]

        # Moving parameters:
        k_ = parameters[9]
        n_ = parameters[10]
        
        # Synergy model parameter:
        gamma = parameters[11]
        
        # S
        tau = (1-np.exp(-uo*y[0]))/uo
        tau_ = (1-np.exp(-uo_*y[1]))/uo_
        theta = (n*tau+n_*tau_)**2/4
        tausy = (1+gamma*n*n_*tau*tau_/theta)
        taut = (n*tau+n_*tau_)
        tautN = taut*tausy
        S = np.exp(-alpha/tautN)
        dtStaut = alpha*np.exp(-alpha/tautN)/tautN**2
        
        dttau = np.exp(-uo*y[0])
        dtautaut = -4*(2*gamma*n**2*n_*tau*tau_/(n*tau + n_*tau_)**3 - gamma*n*n_*tau_/(n*tau + n_*tau_)**2)*(n*tau + n_*tau_) + (4*gamma*n*n_*tau*tau_/(n*tau + n_*tau_)**2 + 1)*n
        dtS = dtStaut*dttau*dtautaut

        dttau_ = np.exp(-uo_*y[1])
        dtautaut_ = -4*(2*gamma*n*n_**2*tau*tau_/(n*tau + n_*tau_)**3 - gamma*n*n_*tau/(n*tau + n_*tau_)**2)*(n*tau + n_*tau_) + (4*gamma*n*n_*tau*tau_/(n*tau + n_*tau_)**2 + 1)*n_
        dtS_ = dtStaut*dttau_*dtautaut_

        # s
        s = np.exp(-uo*y[0])
        dts = -uo*np.exp(-uo*y[0])

        s_ = np.exp(-uo_*y[1])
        dts_ = -uo_*np.exp(-uo_*y[1])

        # p
        A = n*r*(1-np.exp(-uo*y[0]))*k/k_ - n_*(1-np.exp(-uo_*y[1]))
        ro = r*k*ui/(k_*ui_) + (M*M*A/(2*ui_*ui_))*(A+np.sqrt(A*A+4*ui*ui_*r*k/(M*M*k_))) # +- sqrt

        a = M*n*ro**(1/2)
        a_ = M*n_*ro**(-1/2)
        p = a/(a+ui)
        p_ = a_/(a_+ui_)
        
        ## ASR calculus
        #Li = 1/(a*(1-np.exp(-uo*y[0]))+ui)
        #Lo = a(1+uo*y[0]-np.e(uo*y[0])(1-uo*y[0]))/((ui+a(1-np.exp(uo*y[0])))uo)
        
        #Li_ = 1/(a*(1-np.exp(-uo_*y[1]))+ui_)
        #Lo_ = a(1+uo_*y[1]-np.e(uo_*y[1])(1-uo_*y[1]))/((ui_+a(1-np.exp(uo_*y[1])))uo_)
        
        #ra = r((Li_+Lo_)/k_ + (k_-1)/(k_*ui_))/((Li+Lo)/k + (k-1)/(k*ui))


        # Equations of motion
        dT = dtS/S + p*dts/(1-p*s)
        dT_ = dtS_/S + p_*dts_/(1-p_*s_)

        return[dT,dT_]
    
    # Solution time
    iter_time = [x*dt for x in range(step)]
    
    # Solution object and solve
    J = ode_solver()
    J.function = model
    J.ode_solve(y_0=initial_state, t_span=iter_time, params=parameters)
    sol = J.solution
    
    # Take solutions
    for i in range(len(sol)):
        time.append(sol[i][0])
        phi0_t.append(sol[i][1][0])
        phi1_t.append(sol[i][1][1])
    
    phi_t = [np.array(phi0_t), np.array(phi1_t)]
    
#    x = np.zeros(len(phi_t[0]))
#    for i in range(len(phi_t[0])):
#        if phi_t[0][i] < 0 or phi_t[1][i] < 0:
#            x[i] = 1
#    phi_t[0] = phi_t[0][x.astype('bool')]
#    phi_t[1] = phi_t[1][x.astype('bool')]
    
    return np.array(time), phi_t, J
```

```python
### Solution 1: n_ = 1, k_ = 1

#Moving parameters:
k_ = 1
n_ = 1
gamma = 0

parameters = [uo, ui, uo_, ui_, k, n, M, r, alpha, k_, n_, gamma]

# Initial Values
To = []
for i in range(5):
    for j in range(5):
        To.append([i/10,j/10])
del To[0]



# Capture solutions
solutions = []
for i in range(len(To)):
    time, X, _ = Deterministic(steps, 0.01, parameters, To[i])
    solutions.append([time, X])
    
    
# Plot graphic    
color = np.arange(steps)
for i in range(len(To)):
    plt.scatter(solutions[i][1][0],solutions[i][1][1], c=color, cmap='viridis')
    plt.xlim((0,0.4))
    plt.ylim((0,0.4))
    #plt.plot(T[-1], T_[-1], 'D',c='green',markersize=10)
    #plt.plot(T[0], T_[0], '+',c='green',markersize=20)
plt.plot(np.arange(0,1,0.1),np.arange(0,1,0.1), '--', linewidth=4, c='r')
plt.title('Fig 1.a: Simple, k_ = 1, n_ = 1', fontsize=20)
plt.xlabel('T', fontsize=20)
plt.ylabel('T_', fontsize=20)
plt.colorbar()
plt.savefig('a1C.png')
```
