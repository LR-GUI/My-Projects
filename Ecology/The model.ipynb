# Variables

Male vars are with underscore ($u_i$ is female and $\tilde{u_i}$ is male).
Two states of individuals: time-in (i), seeking mate; time-out (o), providing care.

$T$ = care duration, intended duration of time-out. It's the evolutionary trait considered

$n$ = mating group size, number of individuals per breeding attempt. monandry: $n = \tilde{n} = 1$. polyandry: $n = 1, \tilde{n} > 1$.

$k$ = strength of sexual selection; to account for intra/inter sexual selection. separates individuals of a sex into two subpopulations, $1/k$ of them, at maturation, is able to mate and then care, and the rest is not. The subpopulation that is able to mate is called qualified to mate.

$u_i$, $u_o$ = mortality rates during time-in and time-out.

$x$ = competitiveness, models the investment in traits for fitness that decrease longevity.

$r$ = maturation sex ratio (MSR), individuals entering time-in for the first time.

$M$ = mate search coefficient, account for species specific factor such as population density and movement capacity. (the constant factor of the mating rate $a$)

$\alpha$ = shape coefficient of brood survival function, modulates survival acting as a care effectiveness

$\gamma$ = synergy coefficient, modulates increase in effect of care when both parents are present.

$\tau$ = average care duration per time-out (considering deaths).

$a$ = mating rate (of individuals qualified to mate)

$p$ = probability to survive a given time-in period, $p = a/(a+u_i)$

$s$ = probability to survive a given time-out period, $s = e^{-Tu_o}$

$b$ = species brood size, how many individuals on average come from a mate.

$\tau_{total}$ = total care (to be changed $\tau_{total}$ => $C$)

$S$ = brood survival, probability of offspring come to maturation and time-in. $S(\tau_{total}) = e^{-\alpha/\tau_{total}}$

$W$ = average fitness of individuals

$r_o$ = operational sex ratio (OSR), ratio of qualified individuals in time-in, in the mating pool.

$r_a$ = adult sex ratio (ASR), all alive matured individuals, in time-in and time-out 

# The model

Males and females are considered as 2 different populations that are tied together by mating and they coevolve (here, the trait of total duration of care given to each offspring). All we'll say about females is the same for males. Individuals maturate folowing MSR, $r$. After that, they enter time-in for the first time, the mating serch mode. There, females can die at a rate $\mu_i$. But a fraction $1/k$ of them can also mate with rate $a$ (the others just stay there until they die). Here, $k$ is a selection strength, selecting a portion of the population to be qualified for mating and the others to be disqualified. Matings occur in the form of mating groups of $n$ females and $\tilde{n}$ males (but only 1 male and 1 female from each mating group will effectively reproduce). Once they mate, they enter time-out, the caring mode (everyone that mates care, even those that are not actual parents of the offspring). There, females can die at a rate $\mu_o$. But, when they reach a time $T$ spent at time-out without dying, they go back to time-in.

A generation is considered to be all time-ins/time-outs individuals engage until they are all dead (note that there is no senescence, the dying rates stay the same for the whole life). Then, their offspring will make the next generation. The evolution of the trait $T$ is given by the fitness equation feeded with a brood survival activation function, $S$, that depends on the total average contribution of care of both populations together, $\tau{total}$.

Note that since, in the end, each step of the model's dynamics is the result of a whole generation, every variable that is dynamical inside a generation is averaged out. As an example, consider the duration of care given to each offspring. It varies for each offspring, sometimes it is the complete duration $T$, but sometimes it can be any $t<T$ if the individual dies during time-out. But, in the end, the evolution model only considers the average time of care, $\tau$. Also, variables like the ASR and the OSR are averaged over the entire generation.

## Fisher condition:

At first, the mating rates $a$ could be a free parameter. But, in fact, there is a consistency condition they follow. Suppose that we have equal number of operational males and females, $r_o$ and that the mating group sizes, $n$ and $\tilde{n}$, are 1. If we pair every female with a male no one is left without a mate, so they mate at equal rates, $a/\tilde{a}=1$. Now, suppose that, for each operational female, there are 3 operational males, $r_o=3$. If we group every female with a male, we'll still have $2N^o$ males waiting, without a mate (double of females). This decreases the rate in which each male mates. So during each mating event, for 1 female mating, there are 3 males from which only 1 mates and the rest wait, $a/\tilde{a}=3$. Now, if we have mating group sizes of 1 female and 2 males, $n=1$ and $\tilde{n}=2$, only 1 male for each female stays without a mate. For each mating event, there is 1 female mating for each group of 3 males from which 2 are mating and the other is waiting, $a/\tilde{a}=3/2$. Now, if each group has 2 females too, $n=2$, then we are back to a situation of $a/\tilde{a}=3$. For each 2 females mating, we have a group of 6 males from which 2 are mating and 4 are waiting.

From the scenarios above, we see that the ratio of the mating rates, weighted by the relative group size, $n/\tilde{n}$, must be the OSR (the ratio of individuals to be considered in the mating pool, those subject to mating with the rates $a$,$\tilde{a}$):

$\frac{a}{\tilde{a}} = \frac{n}{\tilde{n}}r_o$.

If nothing else affects mating rates, we have $a \propto nr_o^g$ and $\tilde{a} \propto \tilde{n}r_o^{g-1}$. Now, we require that the dependency must be symmetric between females and males, so $g=1/2$ is the only possibility. Let's see how. Symmetry is achieved when the expressions for $a$ and $\tilde{a}$ have the same functional form. Since the numbers of males and females are assymetric in the $r_o=\tilde{N}/N$, only one value of $g$ is able to cause the symmetry. We have

$a \propto r_o^g = \tilde{N}^gN^{-g}$

and

$\tilde{a} \propto r_o^{g-1} = \tilde{N}^{g-1}N^{1-g}$.

For symmetry, we must have the exponent of number of males in the male equation, for $\tilde{a}$, match the exponent of number of females in the female equation, $a$. => $g-1=-g$ => $g=1/2$. The same goes the other way, for the exponent of females in $\tilde{a}$ and males in $a$, giving the same, $1-g=g$ => $g=1/2$, so it's consistent.

Then, we take the proportionality constant and transform it into a species-dependant parameter, in this case the mating search coefficient $M$. => $a=Mnr_o^{1/2}$ and $\tilde{a}=M\tilde{n}r_o^{-1/2}$.

The authors call this the Fisher condition, but it seems to be just a consistency condition for mating rates. Precisely, the Fisher condition comes after mating, determining that (since, irrespective of group sizes, only 1 female and 1 male are able to reproduce and increase their fitness, equally) the total amount of offspring of each sex is the same. So the total amount of offspring an individual has is inversely proportional to the operational sex ratio (if they are qualified for mating, otherwise it is just 0). But note that this is true when we consider that, irrespective of mating group sizes, only 1 male and 1 female are effectively gaining fitness. In the model, this is applied when we consider the condition on the rates above together with the step of dividing $W$ by the group size, $n$, in the fitness equation. A consequence of this implementation that is also a part of what is called the Fisher condition is that each offspring has an equal chance of being male or female. But the crucial hypothesis is that a reproduction event is composed by 1 female and 1 male. If we violate this, considering for example that everyone in the mating group reproduces, the condition on the rates above stays the same, but then we would not divide $W$ by $n$. 


## $S$, brood survival activation function:

$S$ comes from an averaging simplification. The chance of brood survival depends only on the total amount of average care given to the brood, $\tau_{total}$. This leaves out the more structured possibility of this example: an individual mates succesfully 3 times, so the first 2 broods receive the total amount of care $T$. But the last brood may have less amount of care due to the individual's death during care. With this simplification, however, all 3 broods from that individual receive the average care, and their survival is modulated by this number alone (note that $\tau \to T$ when $\mu_o \to 0$). So, regardless of how many broods individuals have in their lifetimes, we consider only their "chosen" duration of care $T$ and their possibility of dying during care to then calculate the total value of their care that is felt by the brood as the probability of their survival. So, how much you want and is able to care translates to your brood as how probable the brood's survival is depending on $\tau$ (via the parents joint variable $\tau_{total}$).
$S$ is chosen so as to be monotonic, of diminishing returns, and 0 for 0 care / 1 for infinite care.
=> $S = e^{-\alpha/\tau_{total}}$, with $\alpha$ being a mathematical shape coefficient that modulates the effectiveness of $\tau{total}$.

About $\tau_{total}$. The model interprets that every one that mates has to go time-out and care for the offspring, even if they are not the actual parents. So, suppose $\tilde{n}=2$, meaning that on average 2 males engage in each (succesfull) breeding attempt. By the model, these 2 males go time-out together and the offspring receive the benefits of care from both. That's why $\tau_{total}=n\tau+\tilde{n}\tilde{\tau}$. This scenario also says that, with higher poliandry, male care as a trait has a higher value (ofspring benefit more from $\tilde{T}$ increasing than $T$), because we have $\tilde{n}$ individuals caring more instead of 1.


## $\tau$, average time of care:

We must calculate the average time of care of an individual considering the possibility of death during time-out, happening with rate $\mu_o$. Here is what we want:

$\tau = P_TT + \int_0^T Q_t t dt$

$P_T$ is the probability of surviving the whole time-out and choosing to leave. $Q_tdt$ is the probability of managing to survive until time $t$ and then dying at time $t$. So we read the equation for $\tau$ as: it is $T$ if we manage to survive, and it is $t$ if we die at $t$, considering all possible $t$'s. So, we take each condition's probability and average out with each condition's time.

$P_T = 1 - \int_0^T e^{-\mu_ot}\mu_odt$, probability of not dying until $T$ is 1 minus probability of dying at any time until $T$. This probability of dying (the integrand) is just the Poisson next-reaction probability with rate $\mu_o$.

$Q_tdt = e^{-\mu_ot}\mu_odt$, probability of dying exactly at $t$ is just the Poisson next-reaction probability, the previous integrand.

Computing everything, we arrive at => $\tau=\frac{(1-e^{-\mu_oT})}{\mu_o}$

The way calculated in the paper is different, it is more efficient but less clear. $\tau$ is $T$ minus the discounts promoted by death. The discounts are $(T-t)$ taken with probability equal to that of dying exactly at $t$. So, when you die at $t$ you lose the remaining time of care, $(T-t)$ (Their way is less clear because we would have to formalize a little more the mathematical justification of discounting time in this way. Before anything, $\tau$ is an average).


## Trait evolution:

Quantitative genetic model - Traits are affected by many genes at many loci, so we consider continuous values of traits (like human height). Complete heritability is commonly assumed (no environmental influences on traits). Fitness is a product of viability and fertility selections. So fitness is a measure of an individual's chances of survival and reproduction, these usually depends on the individual's trait and the mean of the population (measuring how good/bad it goes on average).

Weak selection - A common approximation and modeling tool. It assumes small trait variance, so that individual's traits are always close to the population mean. In the case of this model, we can assume small variance and even constant variance as hypothesis, for our trait is duration of care, $T$, and the whole population of females/males can evolve together (the focus of the model is in differences of traits between sexes, not traits between the same sex, the same population); and being like this, the variation couldn't change much and we assume no forces driving variation. Let's assume a large population of size $N$ with a continuous valued trait $x$. We have $N = \sum_xn(x)$ and:

$\bar{x}=\frac{1}{N}\sum_xxn(x)$ and $var(x)=\frac{1}{N}\sum_x(x-\bar{x})^2n(x)$.

The fitness is the function modulating the fraction individuals with trait of value $x$ in the next generation, $x'$, so $n'=Wn$:

$\bar{x'}= \frac{1}{\sum_{x'}n'(x')}\sum_{x'}x'n'(x') = \frac{1}{\sum_{x'}W(x')n(x')}\sum_{x'}x'W(x')n(x') = \frac{\sum_x xn(x)W(x)}{\sum_x n(x)W(x)}$.

Note that $\sum_{x'}$ is the same as $\sum_x$, because the set of possible values of the trait doesn't change between generations. Then, the differential equation for $x$ becomes:

$\bar{x'}-\bar{x}=dx=\frac{\sum(x-\bar{x})n(x)W(x)}{\sum n(x)W(x)}$.

We can then, using the weak selection assumption, expand $W(x)$ around the mean $\bar{x}$ and arrive at:

$dx \approx var(x)\frac{\partial lnW(\bar{x})}{\partial x}$. => Evolution of a trait expressed as a selection gradient.

In connection with our model (applying on $T$ and considering constant variance), we end up with:

$dT \propto \frac{1}{W}\frac{dW(T)}{dT}$. As seen in the text.


## Fitness equation:

Fitness is the modulator of species change by each generation: $n_x' = W(x)n_x$.
Now, suppose that each individual, in a generation, has on average $W$ offspring. See that, for each trait, $W$ is the same for all individuals. All offspring of all individuals of a trait compose the entire next generation, $n_x'$. But, summing the $W$ offspring of all the $n_x$ individuals, we get $W(x)n_x$. So here, the fitness is the same as the population brood size, for each trait, of this generation. In this model, the entire population holds the same value for the trait $T$, and males and females are treated as different populations coevolving. The trait, and the whole population together, evolves following the shape of the fitness function, along its gradient, as the trait evolution equation says.

So we have to determine the average brood size for our model. The calculus is for females, but is the same for males. Let's go step by step.
First, only $1/k$ of the population ends up mating, by the definition of $k$. The rest has zero offspring. Also, an individual increases its fitness only in $1/n$ of their mates (those for which they are actual parents on average). So:

$W = \frac{1}{nk}Brood + (1-\frac{1}{k})\frac{1}{n}Zero = \frac{1}{nk}Brood$

Here, brood means all the offspring an individual had, not just of one individual brood. We construct the brood as follows: Take individual-brood size (now number of offspring per individual-brood), $b$. Take the average number of mates (the number of individual-broods), $N_m$. Take the brood survival, the chance of every offspring to mature, dependant on the care given, $S$. The summed number of offspring of every mate of that individual, $bN_m$, multiplied by their chance of survival, gives the fraction of offspring that mature and form the brood of that individual, so:

$W = \frac{1}{nk}bN_mS$

Now, what's $N_m$? We have to calculate it. More steps. How many mates we can expect from each (apt) individual, on average? Well, they go time in, if they don't die, they mate. They go time out, if they don't die, go back to time in. So on. Note that they always go time-out when they mate, even if they end up not being the parents (possible for $n>1$). Let's then talk in terms of probability of surviving a period of time in, $p$, and the probability of surviving a period of time out, $s$. So the first mate happens with probability $p$, after that the first return to time in happens with probability $s$. What is the probability of an individual to mate and come back to time in? It's $sp$. What is the probability of mating exactly 2 times? You mate, $p$, then you come back from timeout and mate again, $sp$, then you die in any way before mating again, $1-sp$. So the probability is $psp(1-sp)$. Let's do a consistency check on this dying probability. You can die in two independent forms; during the time out, $1-s$, or survive the time out and die during time in, $s(1-p)$. Adding up the two, you have $1-sp$, the probability of not mating again. If you then add the remaining option, survive the two, $sp$, you have 1. Now, probability of mating exactly 3 times: mate, come back, mate, come back, mate, die - $p.sp.sp.(1-sp) = p(sp)^2(1-sp)$. Now, mate $i$ times: $ Pr_i = p(sp)^{i-1}(1-sp)$. We have what we need for the average number of mates per individual; zero with probability of zero mates, one with probability of 1 mate, 2 with probability of 2 mates, etc:

$N_m = \sum_i iPr_i = (1-p).zero + \sum_{i=1}^{\infty}ip(sp)^{i-1}(1-sp) = p(1-sp)\frac{\partial}{\partial (sp)}\sum_{i=1}^{\infty}(sp)^i = p(1-sp)\frac{\partial}{\partial (sp)}\frac{sp}{1-sp} = \frac{p}{1-sp}$

Again, the expression shown in the paper, without any explanation, is impossible to justify as it is, but it's an intermediate step in the solution. I suspect that they put it there without realizing or caring for this. As a curiosity, we can build a much clearer expression using an important result for averages of infinite positive integer random variables:

$E(x) = \sum_i iP(x= i) = \sum_i P(x\geq i)$ => $N_m = \sum_i P(n_{mates}\geq i) = \sum_{i=0}^{\infty} p(sp)^i = \frac{p}{1-sp}$

To compute the probability of $i$ or more mates, we just drop off the death clause. Here is what we have now:

$W = \frac{1}{nk}\frac{p}{1-sp}bS$

Note that the authors define $b$ in a slightly different manner. Theirs is the female individual-brood size already accounting for $n$, so the male counterpart is $\tilde{b}=\frac{n}{\tilde{n}}b$, so the fitness equations are slightly different. Now, this is interesting: any constant scaling of $W$ is irrelevant for trait evolution, so dividing $W$ by $n$ has no effect in the results (as for the parameter $b$ itself).

Ok. One more step to go: What are $p$ and $s$ in terms of model parameters? 

So, variables $p$ and $s$ are the survival probabilities of time-in and time-out, the probability of ending the period without dying. We have calculated $s=P_T$ above, so => $s=e^{-\mu_oT}$.

Now, for $p$, we have two possible "reactions" that exit time-in, mating with rate $a$ and dying with rate $\mu_i$. Then, whenever time-in ends, the chances of survival are the chances of a mating "reaction" to occur => $p=\frac{a}{a+\mu_i}$ (likewise, the probability of a death "reaction" would be $1-p = \frac{\mu_i}{a+\mu_i}$).

Now, finally, let's derive the fitness equation and connect it with the trait evolution equation:

$dT \propto \frac{1}{W}\frac{dW}{dT} = \frac{1}{S}\frac{dS}{dT} + \frac{p}{1-ps}\frac{ds}{dT}$.

Note that we are considering $a$ as a constant in the derivation, but it isn't (it depends on $r_o$, that depends on $T$). We approximate it as a constant with the assumption that mating rates remain practically constant for each step in $T$. For consistency, we would have to verify that, linking $da$ to $dT$ and seeing how the change depends on $T$ (O(T)? O(1/T)?, etc)

Now, how does these derivatives connect to model parameters? $S$, for example, is a function of $\tau_{total}$, which is a function of $\tau$, which finally is a function of $T$. We go from $S$ to $T$ using the chain rule:

$\frac{dS}{dT} = \frac{dS}{d\tau_{tot}}\frac{d\tau_{tot}}{d\tau}\frac{d\tau}{dT}$.

The derivative between $\tau_{total}$ and $\tau$ is just $n$, but the authors have forgot to include it in their model. The text after equation 3 in the paper makes this very explicit. This is a minor problem, because $n$ is mostly 1 anyway. It becomes a bigger problem in the synergistic model, where this derivative is much uglier.


## Operational and adult sex ratios


There is now only one thing in our way to model the trait evolution; we don't know the $r_o$ that appears in our derivation of the mating rate $a$. Also, we don't know the adult sex ratio, $r_a$, that is not needed for the model, but needed to make the analysis of ASR dependency. They are similar and we will write them in terms of the average total times spent in time-in and time-out by each qualified individual during their lifetime.

The ASR and the OSR are numbers associated with the joint population of males and females. They are the average ratios along the whole generation. The ratio OSR is the same as a joint population that has $r_o$ qualified males in time-in for each 1 qualified female in time-in, i.e., $\tilde{N}^{in}/\tilde{k}=r_oN^{in}/k$, for the entire generation. The MSR, $r$, is the same as the joint population having $r$ males ready to enter time-in for the first time for each 1 female in this situation, i.e., $\tilde{N}^m=rN^m$. Now, the ASR is there to measure the imbalance that comes from deaths after that, deaths during time-in and time-out. It is the same as a joint population in which there are $r_a$ males both in time-in and time-out for each female in the same situation, i.e. $\tilde{N}^{in+out}=r_aN^{in+out}$, for the entire generation.

Note the connection between them: The MSR measures the average ratio of individuals coming to engage in time-in and time-out (in reality, this number varies as new generations are created). The ASR measures the average ratio of these individuals while they are in time-in and time-out (in reality, this number varies as individuals die during time-in and time-out). The OSR measures the average ratio these individuals only considering those in time-in that are qualified to mate (in reality, this number varies as individuals die during time-in, go to time-out, or come from time-out).

So let's talk about the variables we'll use to compute the ASR and the OSR. We define $L_i$ as the average total time spent by 1 qualified female during time-in, counting all time-ins of that female. And $L_o$ is the average total time spent by 1 qualified female during time-out, counting all her time-outs. For the population of males, we have $\tilde{L_i}$ and $\tilde{L_o}$. We choose these variables because they can be extracted from our parameters, specifically we use the reaction rates for leaving time-in and time-out, given in terms of $T$, $a$, $\mu_i$, and $\mu_out$ (there is a catch here, but this is the reasoning).

How the ASR and OSR relate to these $L$s? The following is valid separately for qualified and unqualified individuals: We consider the total time spent during time-in (-out) by each individual as a weight of that individual's participation in time-in (-out) during its generation. By that, we can say that, the sum of each individual that matures and participates in the generation, weighted by the total time they spend in time-in (-out), is equal to its participation in time-in (-out). For calculating this participation, in turn, the weighted sum is the same as the total number of mature individuals times the average time an individual spends during all their time-ins (-outs), the $L$s. Then, the ratio between the total participation of males and of females in time-in (-out), on average, is the same as the average ratio between actual individuals during time-in (-out). These equalities only hold on average. Here is what it all means, using time-in:

$\frac{MaleParticipation^{in}*}{FemaleParticipation^{in}*} = \frac{\sum_{j=1}^{\tilde{N}^m*}\tilde{L}_i^j}{\sum_{j=1}^{N^m*}L_i^j} = \frac{(\tilde{N}^m\tilde{L_i})*}{(N^mL_i)*} = \frac{\tilde{N}^{in}*}{N^{in}*}$

The * terms mean qualified individuals, unqualified individuals, or their sum. 

Note that the OSR is the last term above when only qualified individuals are considered. So:

$r_o = \frac{(\tilde{N}^m\tilde{L_i})(qualif)}{(N^mL_i)(qualif)} = \frac{k\tilde{N}^m\tilde{L_i}}{\tilde{k}N^mL_i}=r\frac{k\tilde{L_i}}{\tilde{k}L_i}$.

The ASR is the same, but considering everyone and both time-in and time-out. Before laying out its expression, let's calculate the $L$s.




## Total individual times, $L_i$ and $L_o$

We calculate for females and it is analogous for males. $L_i$, the average total time in of an individual, can be broken into the average time of a single time-in times the average number an individual enters time-in during their life. So:

$L_i = n_{in}t_{in}$

Once in time-in, a "reaction" to exit is given at a constant rate, either by mating, with rate $a$, or dying, with rate $\mu_i$. It follows a poissionian distribution, the first-reaction time is the inverse of the combined rate of exiting:

$t_{in} = \frac{1}{a+\mu_i}$.

If the individual is not qualified to mate, their mating rate is zero so that $t_{in} = \frac{1}{\mu_i}$ for them. Also their $n_{in}=1$, because they never leave time-in. So, for these individuals, the average time spent during their only time-in is $\frac{1}{\mu_i}$. Following this, we consider only qualified individuals.

The average number of time-ins is defined as the sum of number of time-ins times the probability of having exactly that number of time-ins, for all possibilities. The probability of having 1 time-in is 1 (here we are talking about an already mature individual). Now, the probability of returning to time-in, after exiting, we calculated above as $sp$. Let's here use the trick of the average mentioned above (the calculus here is pretty much the same as that one) and make it the sum of the probabilities of having that much or more time-ins:

$n_{in} = \sum_{i=1}^{\infty}P(N_{in}\geq i)$.

That probability is the same as the chance of returning to time-in $i-1$ times, $(sp)^{i-1}$ (if you return once, you got 2 time-ins so far). Then:
$n_{in} = \sum_{i=1}^{\infty}(sp)^{i-1} = \frac{1}{1-sp}$.

Now, we substitute $s$ and $p$ and get:

$L_i = \frac{1}{\mu_i+a(1-e^{-\mu_o T})}$.

Now, for $L_o$ we do the same:

$L_o = n_{out}t_{out}$.

If the individual is not qualified, it has zero average time-out time, because it never goes to time-out.

For qualified individuals, we find $n_{out}$ with the same reasoning, only considering that the first time-out isn't certain, it comes with probability $p$ (that of the first mate). So we have:

$n_{in} = p\sum_{i=1}^{\infty}(sp)^{i-1} = \frac{p}{1-sp}$.

Now, the average time of a time-out is just $\tau=\frac{(1-e^{-\mu_oT})}{\mu_o}$! So we have:

$L_o = \frac{p\tau}{1-sp} = \frac{a(1-e^{-\mu_oT})}{\mu_o(\mu_i+a(1-e^{-\mu_oT}))}$.

This expression is different than the one from Kokko and Jennions that is reproduced in this paper. Unfortunately, they commited a sign error that considerably affects the dependancy on $T$. Consequently, their ASR is wrong through $L_o$. (Their method of derivation of $L_o$ is different from the one above, but it is equivalent, so their equations A3 and A4 are valid. The first equality of A5 is also valid, but the second contains the error).

So knowing also the $L$s for unqualified individuals, we can calculate the ASR the same way we did for the OSR above:

$r_a = \frac{(\tilde{N}^m\tilde{L_i})(qualif+unqualif)+(\tilde{N}^m\tilde{L_o})(qualif+unqualif)}{(N^mL_i)(qualif+unqualif)+(N^mL_o)(qualif+unqualif)} = \frac{\frac{1}{\tilde{k}}\tilde{N}^m\tilde{L_i}+\frac{\tilde{k}-1}{\tilde{k}}\tilde{N}^m\frac{1}{\tilde{\mu_i}} + \frac{1}{\tilde{k}}\tilde{N}^m\tilde{L_o}+\frac{\tilde{k}-1}{\tilde{k}}\tilde{N}^m.zero}{\frac{1}{k}N^mL_i+\frac{k-1}{k}N^m\frac{1}{\mu_i} + \frac{1}{k}N^mL_o+\frac{k-1}{k}N^m.zero} = r\frac{\frac{1}{\tilde{k}}(\tilde{L_i}+\tilde{L_o})+\frac{\tilde{k}-1}{\tilde{k}}\frac{1}{\tilde{\mu_i}}}{\frac{1}{k}(L_i+L_o)+\frac{k-1}{k}\frac{1}{\mu_i}}$.

Knowing the $L$s, we can compute the $r_a$ by this expression.

Remembering the OSR, we have:

$r_o = r\frac{\tilde{kL_i}}{\tilde{k}L_i}$.

Now, we can't obtain the $L$s without already having the OSR. Note that $a$ is not an actual parameter of the model. It is dependant on $r_o$ itself, so the equation for $r_o$ is implicitly defined. We must obtain $r_o$ first, then the $L$s, then $r_a$. Using $a=Mnr_o^{1/2}$:

## Synergistic model

The synergy of care is an additional gain in effectiveness of care when the care was given by both parents simultaneously. Effectiveness of care in the model is measured as a function of total duration of care $\tau_{total}$. With this, we can define a density of care, the variable representing units of care, as the time spent caring. This is made explicit by the definition of $\tau_{total}=n\tau+\tilde{n}\tilde{\tau}$. The density of care for females is the total time females spend caring, $\rho = n\tau$; the same for males, $\tilde{\rho} = \tilde{n}\tilde{\tau}$. This variable represents a weight of care, the more you have of it, the more effective is care (in terms of making $S(\tau_{total})$ higher). But it is also the total time of care for each population. Now, if this time is spent together, there is synergy, and the gain should be higher than for the fraction of this time that is spent alone.

Now, suppose that females care for less time, so $\rho<\tilde{\rho}$. Define the positive difference in time of care $\delta = \tilde{\rho}-\rho$. Then, for a period $\rho$, both are caring, so there is synergy. For a period $\delta$, only males are caring, without synergy.

The authors consider a synergy modulator $\gamma$ so that 1 unit of care time with synergy has the same density as $1+\gamma$ units of care time with no synergy. So $\rho$ is a time that is $1+\gamma$ times more effective than $\delta$. Without synergy, for $\gamma=0$, we have $\tau_{total}=\rho+\tilde{\rho}=2\rho+\delta$. The new total (density of) care, considering the synergy, is built the same way but considering that $\rho$ is $1+\gamma$ more weighted than $\delta$:

$\tau_{total}^s=2\rho.(1+\gamma)+\delta.1 = 2\rho(1+\gamma)+\delta$

The authors propose a new form of the total duration that is $\tau_{total}^s=\tau_{total}(1+e\gamma)$. What we have to do now is reorganize our expression in this form and discover what is the value of $e$. We reorganize and change back $\delta$ to $\tilde{\rho}$:

$\tau_{total}^s=(\tilde{\rho}+\rho)(1+\frac{\rho}{\frac{\rho+\tilde{\rho}}{2}}\gamma) = \tau_{total}(1+e\gamma)$

So $e=\frac{\rho}{\frac{\rho+\tilde{\rho}}{2}}$. Now, this expression considers that females care for less time. It the males care less, we have to exchange the tildes, and it is another expression. What we can do to unify them is symmetrize the expression, cancelling the differences. If we consider $e=\frac{\rho}{\frac{\rho+\tilde{\rho}}{2}}\frac{\tilde{\rho}}{\frac{\rho+\tilde{\rho}}{2}}$, the expression maintains itself between the other two and is unified. This increases a little the effectiveness of the synergy, but has the same behavior.

Now, this is the expression as defined in the paper:

$\tau_{total}^s=\tau_{total}(1+\frac{\gamma\rho \tilde{\rho}}{\theta})$.

For $\theta=n\tilde{n}(\frac{\tau+\tilde{\tau}}{2})^2$. But, in the way we reasoned here, we should have:

$\theta=(\frac{\rho+\tilde{\rho}}{2})^2 = (\frac{n\tau+\tilde{n}\tilde{\tau}}{2})^2$.

This is another alteration that doesn't change much, but it can sensibly affect the dependency of the model on $n$ and $\tilde{n}$, the numbers of individuals participating in each breed.

## Questions to the authors

- The expression for total care dutation is $\tau_{total} = n\tau + \tilde{n}\tilde{\tau}$. This considers that all individuals that participate in breeding attempts go to time-out, not just 1 of them (and the rest of the model is consistent with this, e.g. $p$ considers that everybody who mates go to time-out). So for $\tilde{n}>1$ each brood has more than one caring male parent, on average. Is this correct?

- The parameter $n$ has 4 roles in the model: 1) It enters in the equation for $a$, saying that everybory in the mating group goes time-out, not just a couple; 2) It enters in $\tau_{total}$, saying that everyone from the mating group stays in time-out caring for the offspring together, irrespective of who are the actual parents; 3) It enters in $W$, saying that only the actual parents increase their fitness... Now, 1 and 2 are strange, and 3 is left as the core role of parental uncertainty in evolution of care. But 3 doesn't affect the results, because it's a scaling of $W$.

- For the simple model, the chain rule for calculating $\frac{dS}{dT}$ doesn't consider $\frac{d\tau_{total}}{d\tau}$, which is a multiplicative factor of $n$. Why?

- For the synergistic model, that same derivative is also strange. It now considers $\frac{d\tau_{total}}{d\tau}$, but the final answer also lacks a mutiplicative factor of $n$ (which maintains the consistency with the simple model when $\gamma=0$). And also $\frac{d\theta}{d\tau}$ is considered to be zero. Is this correct?

- Now, about $\theta$, why is it $\theta = n\frac{\tau+\tilde{\tau}}{2}\tilde{n}\frac{\tau+\tilde{\tau}}{2} = n\tilde{n}\frac{(\tau+\tilde{\tau})^2}{4}$ and not $\theta = \frac{(n\tau+\tilde{n}\tilde{\tau})^2}{4}$? The later would maintain consistency in what is the currency of care effectiveness, which is $n\tau$.

Old Trivers model: Initial differences feedbacking to greater differences. Initial anisogamy leads to further sex differentiation. Since the female cares more from the beginning, she keeps caring more. The same for male competing, it feedbacks into more competition.
But this is a verbal model. These claims must be carefully confirmed by evidence. Mathematical models are a proposed way to provide evidence from mechanistic considerations.

Kokko and Jennions previously devised a model for that purpose. Their model shows a force in opposition to the amplification of differences: when the operational sex ratio is male-biased, males have no incentive to desert caring, since they will have to compete for mates (strange... it depends on efficiency of caring and competition). They conclude that, apart from other differences, sex roles are stable against anisogamy (not necessarily, even; it seems...). Other differences include: stronger sexual selection on one sex, mixed parentage, biased adult sex ratio).
Argument of this paper: the opposing force conclusion is wrong. With an updated model, the paper corrects the wrong assumptions. The effects of those other factors for sex differences remain similar.
